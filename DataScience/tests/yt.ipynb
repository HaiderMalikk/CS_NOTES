```python
# Apply the formula: The algorithm applies the formula 
# Initial seed 
# X =1 (seed)
# m=100 (modulus, limits the range to 0-99),
# a=37 (multiplier),
# c=10 (increment),
# Formila Xn+1 = (a * Xn + c) mod m
# use SUM = SUM *10 + X! to 
# Accumulate the random values mathematically * 10 so each itr shifts the dec place left by 1 and appends the new randome number This is similar to adding a new digit to a number. 
        # First iteration:
        # X1 = 7 (let's assume X1 is 7 for this example).
        # Multiply SUM by 10: 0 * 10 = 0.
        # Add X1: 0 + 7 = 7. Now SUM = 7.
        # Second iteration:
        # Assume X1 = 3 for this round.
        # Multiply SUM by 10: 7 * 10 = 70.
        # Add X1: 70 + 3 = 73. Now SUM = 73.
        # Third iteration:
        # Assume X1 = 5.
        # Multiply SUM by 10: 73 * 10 = 730.
        # Add X1: 730 + 5 = 735. Now SUM = 735.

# to show problems:
# if seed odd or even coin head or tail each time 
# cannot make a radnome number to be seed thast like rd number needed to gen another rd number 
# to fix just use time as seed
# then if we exe func in loop time too close and always heads or tails 
# then add time sleep but then its at increments so still always head ot tails, to fix multiple unix time to get 0.xx digits or use mod again to get a range (better)

# for prob (func range makes num more rand and flips makes prob ore 50 -50 likly)

# show rd number and everything step by step 

import time
#EX:
def coin_flip(seed):
    # Parameters for Linear Congruential Generator
    m = 100  # Modulus
    a = 37   # Multiplier
    c = 10   # Increment
    X = seed  # Initial seed 
    
    # Initialize an integer to store the random number
    random_number = 0
    
    # Generate a random sequence using LCG
    for _ in range(100):  # We can adjust the number of iterations to get more randomness
        X = (a * X + c) % m # formula
        random_number = random_number * 10 + X  # Combine the values as a larger number
    
    # print("Generated Random Number: ", random_number)
    
    # Coin flip based on the random number being even or odd
    if random_number % 2 == 0:
        result = 'head'
    else:
        result = 'tail'
    sleep_time = (random_number % 10 + 1) * 0.01 
    return sleep_time, result  
 # times between 0.1 to 0.5 seconds, which will vary slightly each time. or rd * 0.00000000001
 
# sim n flips 
def test_flips(flips):
    flips = flips
    tail = 0 
    head = 0
    for _ in range(flips):
        sleep, result = coin_flip(time.time_ns()) # seed = time.time_ns() to get a more random number
        time.sleep(sleep)
        if result == 'tail':
            tail+=1
        else:
            head += 1
    
    print(f"heads: {head}, Tails: {tail}")
    
    # Calculate actual probability as percentage
    prob_head = (head / flips) * 100
    prob_tail = (tail / flips) * 100
    
    print(f"Probability of heads: {prob_head}%")
    print(f"Probability of tails: {prob_tail}%")
    
    closeness = abs(prob_head - prob_tail)
    print("diff: ",closeness)
    
    # Check if the probabilities are roughly 50-50 (perfect for a fair coin)
    if abs(prob_head - prob_tail) == 0:
        print("PERFECT")
    else:
        print("Not Perfect")
        
    return prob_head, prob_tail, closeness
    
print("running...")
prob = []
for i in range(1, 31):
    prob.append(test_flips(i*10))
print(prob)


phead = [p[0] for p in prob]
ptail = [p[1] for p in prob]
diff = [p[2] for p in prob]
flips = [x * 10 for x in range(1,31)]

data = list(zip(flips, phead))
print(data)
    
import matplotlib.pyplot as plt
plt.title("Coin Flip Probabilities For Heads")
plt.plot(flips, phead)  
plt.show()

plt.title("Coin Flip Probabilities For Tails")
plt.plot(flips, ptail)
plt.show()
    
plt.title("Coin Flip Probabilities For Difference")
plt.plot(flips, diff)
plt.show()    
```
